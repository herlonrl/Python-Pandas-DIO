```python
#Dicionários
#Para criar dicionários utilizamos as chaves {}
dc = {"Maça":20, "Banana":10, "Laranja":15, "Uva":5} #Dicionários trabalham com o conceito chave e valor
```

```python
dc
```

```python
# Acessando o valor de um dicionário através da chave
dc["Maça"]
```

```python
# Atualizando o valor da Maçã
dc["Maça"] = 25
dc
```

```python
#Retorna todas as chaves do dicionário
dc.keys()
```

```python
#Retona todos os valores do dicionário
dc.values()
```

```python
#Verifica  se já existe uma chave no dicionário e caso não exista inserir
dc.setdefault("Limão", 22)
dc
```
